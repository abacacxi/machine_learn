```python
# --- Importação de bibliotecas ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# --- Configurações de exibição ---
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid')

# --- Carregando o dataset ---
df = pd.read_csv('students_performance.csv')
df.head()
df.tail()
```

```python
missing = df.isnull().sum()
missing_percentage = (missing / len(df)) * 100
missing_percentage.sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,6))
missing_percentage.plot(kind='bar')
plt.title("Porcentagem de Valores Faltantes por Coluna")
plt.ylabel("Porcentagem (%)")
plt.show()
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(16, 5 * ((len(num_cols) // 3) + 1)))

for i, col in enumerate(num_cols, 1):
    plt.subplot((len(num_cols) - 1) // 3 + 1, 3, i)
    sns.boxplot(data=df, x=col)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribuição de {col}")
    plt.show()
```

```python
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    plt.figure(figsize=(8,4))
    
    df[col].fillna("Missing").value_counts().plot(kind='bar')
    
    plt.title(f"Distribuição das Categorias: {col}")
    plt.xlabel(col)
    plt.ylabel("Contagem")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
