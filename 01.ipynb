4-Remoção de duplicidade

```python
# --- Verificando duplicatas ---
print(f"Duplicatas encontradas: {df.duplicated().sum()}")

# --- Removendo duplicatas ---
df.drop_duplicates(inplace=True)
print(f"Após remoção: {df.duplicated().sum()}")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_cols = ['gender', 'parental_education', 'extracurricular',
                    'tutoring', 'internet_quality', 'health_status']

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

df.head()
```

```python
# --- Tratando valores infinitos e ausentes ---
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Preenche nulos com a mediana
numeric_cols = ['age', 'study_hours_week', 'attendance_rate', 'sleep_hours',
                'previous_scores', 'family_income', 'final_grade',
                'study_efficiency', 'sleep_quality']

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# --- Normalização com StandardScaler ---
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

df[numeric_cols].head()
```
